Com c√≥digo python podemos de forma automatizada obter as sequencias dessas prote√≠nas usando o UniProt ID:

```python
!pip install biopython
import requests
from Bio import SeqIO
from tempfile import NamedTemporaryFile
import os

def sequence_for_uniprot_id(uniprot_id):
    r = requests.get(f"https://www.uniprot.org/uniprot/{uniprot_id}.fasta")
    r.raise_for_status()

    with NamedTemporaryFile(suffix=".fasta", mode="w", delete=False) as tmp:
        tmp.write(r.text)

    sequence = SeqIO.read(tmp.name, format="fasta")
    os.remove(tmp.name)

    return sequence

uniprot_ids = ['E5RU00', 'E5RTZ2', 'E5RU24', 'E5RU28']

for uniprot_id in uniprot_ids:
    sequence = sequence_for_uniprot_id(uniprot_id)
    print(f"UniProt ID: {uniprot_id}")
    print(f"Description: {sequence.description}")
    print(f"Sequence: {sequence.seq}\n")
```

Depois de termos as sequencias carregadas no python podemos aplicar diferentes bibliotecas para as analisar, como por exemplo fazer o seu alinhamento:

```python
from Bio.Align import PairwiseAligner

sequences = [sequence_for_uniprot_id(uniprot_id) for uniprot_id in uniprot_ids]

aligner = PairwiseAligner()


alignments = aligner.align(sequences[0].seq, sequences[1].seq)
print(f"Alignments for UniProt IDs: {uniprot_ids[0]} and {uniprot_ids[1]}")
for alignment in alignments:
    print(alignment)
    break




```
