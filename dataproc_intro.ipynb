# 🌐 Acesso Programático a Bases de Dados de Bioinformática



🤔 Acesso programático? Porquê? Não é muito mais fácil usar o browser?

## O acesso programático às bases de dados tem várias vantagens:

* **Automatização**: O acesso programático permite a automatização de tarefas repetitivas. Por exemplo, se for necessário extrair os mesmos dados de diferentes partes de uma base de dados, um script pode fazê-lo de forma mais eficiente do que um ser humano utilizando um browser.

* **Integração**: Os dados extraídos podem ser facilmente integrados com outras partes do seu pipeline de análise de dados. Por exemplo, pode extrair dados, processá-los e visualizá-los utilizando o mesmo script.

* **Reprodutibilidade**: O acesso programático garante que o processo de extração de dados é reproduzível. Isto é especialmente importante na ciência, onde a capacidade de reproduzir resultados é fundamental.

* **Personalização**: Com o acesso programático, o utilizador tem mais controlo sobre o que está a fazer. Pode personalizar o script para satisfazer as suas necessidades específicas.

Em resumo, embora a utilização de um browser possa ser mais fácil para tarefas simples e únicas, o acesso programático oferece maior flexibilidade e controlo, especialmente quando se trata de tarefas complexas e repetitivas como tipicamente acontece na Bioinfomática.

# 🖥️ Objectos de dados eficientes e módulos para os trabalhar

Os dados em bioinformática e noutras áreas da computação e ciência de dados são tipicamente organizados de forma eficiente em objectos como **vetores**, **matrizes**, **arrays**, **dataframes**, **listas**:

- **Vetores**: É uma conjunto unidimensional de elementos do mesmo tipo.

Exemplos de Bibliotecas ‘Fundacionais’ 🏗️:

Segue-se uma breve explicação de algumas destas entradas:

`atc_classifications`: Os códigos do sistema de classificação Anatomical Therapeutic Chemical (ATC) para a molécula.

`availability_type`: O tipo de disponibilidade da molécula.

`biotherapeutic`: Informações sobre as propriedades bioterapêuticas da molécula.

`black_box_warning`: Indica se existe um aviso "negativo" para a molécula.

`molecule_chembl_id`: A identificação ChEMBL da molécula.

`molecule_hierarchy`: A hierarquia da molécula na base de dados ChEMBL.

`molecule_structures`: As estruturas da molécula em vários formatos, como SMILES, InChI e molfile.

`molecule_synonyms`: Trata-se de uma lista de sinónimos para o nome da molécula.

# 🧫 A base de dados UniProt

 A UniProt https://www.uniprot.org/ é uma base de dados muito abrangente e popular, de alta qualidade,que permite o acesso gratuito a sequências de proteínas e informações funcionais sobre proteínas de diversas espécies.

 Vamos por exemplo usar o UniProt para procurar sequencias da PfATP6 que foi proposta como alvo da artemisinina.

image.png

Figura 4: Captura de ecran do UniProt mostrando diversas proteínas previstas da pfATP6.

Com código python podemos de forma automatizada obter as sequencias dessas proteínas usando o UniProt ID:

```python
!pip install biopython
import requests
from Bio import SeqIO
from tempfile import NamedTemporaryFile
import os

def sequence_for_uniprot_id(uniprot_id):
    r = requests.get(f"https://www.uniprot.org/uniprot/{uniprot_id}.fasta")
    r.raise_for_status()

    with NamedTemporaryFile(suffix=".fasta", mode="w", delete=False) as tmp:
        tmp.write(r.text)

    sequence = SeqIO.read(tmp.name, format="fasta")
    os.remove(tmp.name)

    return sequence

uniprot_ids = ['E5RU00', 'E5RTZ2', 'E5RU24', 'E5RU28']

for uniprot_id in uniprot_ids:
    sequence = sequence_for_uniprot_id(uniprot_id)
    print(f"UniProt ID: {uniprot_id}")
    print(f"Description: {sequence.description}")
    print(f"Sequence: {sequence.seq}\n")
```

Depois de termos as sequencias carregadas no python podemos aplicar diferentes bibliotecas para as analisar, como por exemplo fazer o seu alinhamento:

```python
from Bio.Align import PairwiseAligner

sequences = [sequence_for_uniprot_id(uniprot_id) for uniprot_id in uniprot_ids]

aligner = PairwiseAligner()


alignments = aligner.align(sequences[0].seq, sequences[1].seq)
print(f"Alignments for UniProt IDs: {uniprot_ids[0]} and {uniprot_ids[1]}")
for alignment in alignments:
    print(alignment)
    break




```
