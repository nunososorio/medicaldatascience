# Data Processing with R: Differential Expression Analysis of Transcriptomic Data Using EdgeR and the Impact of Different Normalization Methods üß¨üî¨üìä

üìà In the field of RNA-seq data analysis, scaling-based normalization methods are commonly used, such as:

1. **TMM (Trimmed Mean of M-values)**: This method estimates scale factors between samples (M-values), which minimize log-fold changes between samples for the majority of genes. TMM is based on the assumption that most genes are not differentially expressed.

2. **RLE (Relative Log Expression)**: This method calculates the scale factor as the median of ratios, for each gene, of its read counts to the geometric mean of read counts for that gene across all samples.

3. **TMMwsp (Weighted Trimmed Mean of M-values with Singleton Pairing)**: This method is a variant of TMM that is designed to work better for data with a high proportion of zeros. In the TMM method, genes that have zero count in any sample are ignored when comparing sample pairs. In the TMMwsp method, the positive counts of such genes are reused to increase the number of features by which samples are compared.

# üåê Programmatic Access to Bioinformatics Databases

ü§î Programmatic access? Why? Isn't it much easier to use a browser?

Programmatic access to databases has several advantages:

*   **Automation:** Programmatic access allows the automation of repetitive tasks. For example, if you need to extract the same data from different parts of a database, a script can do it more efficiently than a human using a browser.
*   **Integration:** Extracted data can be easily integrated with other parts of your data analysis pipeline. For example, you can extract data, process it, and visualize it using the same script.
*   **Reproducibility:** Programmatic access ensures that the data extraction process is reproducible. This is especially important in science, where the ability to reproduce results is fundamental.
*   **Customization:** With programmatic access, the user has more control over what they are doing. You can customize the script to meet your specific needs.

In summary, while using a browser may be easier for simple, one-time tasks, programmatic access offers greater flexibility and control, especially when dealing with complex and repetitive tasks, as is typically the case in Bioinformatics.
